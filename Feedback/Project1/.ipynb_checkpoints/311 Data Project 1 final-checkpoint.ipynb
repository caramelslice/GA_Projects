```python
import pandas as pd
import seaborn as sns
from __future__ import division
```

```python
location = '../../../downloads/311_Service_Requests_from_2010_to_Present.csv'
data_311 = pd.read_csv(location, header=0,parse_dates = ['Created Date'] )
```

```python
data_311['Hour'] = [date.strftime('%H') for date in data_311['Created Date']]
data_311['Day'] = [date.strftime('%d') for date in data_311['Created Date']]
data_311['Month'] = [date.strftime('%m') for date in data_311['Created Date']]
data_311['Year'] = [date.strftime('%Y') for date in data_311['Created Date']]
data_311['Date'] = [date.strftime('%D') for date in data_311['Created Date']]
```

```python
data_311['Total'] = data_311.Agency
```

```python
abc =data_311.groupby('Agency').count()
agencycount = abc['Created Date']
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
agencycount.plot(kind = 'bar', )
plt.xlabel('Agency')
plt.ylabel('Call Volume')
plt.title('Calls per Agency')
```

```python
a_count={}
for row in data_311.index:
   if data_311['Agency'][row] in a_count:
        a_count[data_311['Agency'][row]] += 1
   else:
      a_count[data_311['Agency'][row]] = 1
```

```python
from collections import OrderedDict
from operator import itemgetter
d = OrderedDict(sorted(a_count.items(), key=itemgetter(1), reverse = 1))
d
```

From the above graph and dict it is quite obvious that all agencies don't have equal call volumes. the calls per agency range from 1 call to 1.2 million. So I am only focusing on only the top 10 agencies.

```python
a = ['HPD','DOT','NYPD','DEP','DSNY','DOB','DPR','DOHMH','DOF','TLC',]
for i in a:
    data_311[i] = data_311.apply(lambda row: 1 if row['Agency'] == i else 0, axis = 1)    
```

```python
total = data_311.groupby(['Date']).agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                             'DOHMH':'sum','DOF':'sum','TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
total.plot()
plt.xlabel('Date')
plt.ylabel('Call Volume')
plt.title('Calls Over Time')
```

There is a clear cycle in all the calls, lets first see if that is true of all the agencies and then we will later figure out why we have cyclical behavior.

```python
t = total
del t['Total']
del t['HPD']
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
t.plot()
plt.xlabel('Date')
plt.ylabel('Call Volume')
plt.title('Calls Over Time')
```

So after removing the total and the housing department it is clear that this cycle is evident in all agencies. Lets see why this is.

```python
yearlycount = data_311.groupby(['Year']).agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                              'DOHMH':'sum','DOF':'sum' ,'TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
yearlycount.plot(kind = 'bar')
plt.ylabel('Call Volume')
plt.title('Calls per Year')
```

Overall call volume seems  to be increasing which makes sense as the population is increasing, however the sample size of two years is to small to make many other claims.

```python
monthlycount = data_311.groupby(['Month']).agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                              'DOHMH':'sum','DOF':'sum' ,'TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
monthlycount.plot()
plt.title('Calls per Month')
plt.ylabel('Call Volume')
```

So this shows that there are more calls during the winter especially from the housing department. This is because people call to complain about lack of heat. Calls overall also increase because there are more things people are interested in(ex: Snow plowing, Alternate side parking suspended). Since the other agencies are to small to compare I am going to plot them without the Total and HPD

```python
month = monthlycount
del month['Total']
del month['HPD']
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
month.plot()
plt.ylabel('Call Volume')
plt.title('Calls per Month')
```

This shows some seasonality in other agencies as well.  Calls regarding the Dept. of Transportations happen more in the winter. With a 20,000 call difference in July compared to February. Other Agencies with high sesonality are the NYPD, DEP and the DPR each having more calls during the summer. This makes sense since when the weather gets better people get out more and are more likely going to notice something that involves the Dept. or Parks and Recreations or Dept. of Enviornmental Protection. Also since people are out more crime seems to increase.

```python
dailycount=data_311.groupby(['Day']).agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                              'DOHMH':'sum','DOF':'sum' ,'TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
dailycount.plot()
plt.ylabel('Call Volume')
plt.title('Calls per day')
```

This doesn't necessarily show any trends except for slight spikes on the 8th, 18th and 24th of the month, especially with the housing department. Calls are almost half for the 31st which makes sense because only 7 out of the 12 months have 31 days. It is kind of difficult to look into trends in other agencies so I am going to plot the agencies without the count and housing department

```python
day = dailycount
del day['Total']
del day['HPD']
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
day.plot()
plt.title('Call per Day')
plt.ylabel('Call Volume')
```

I am unable to find many patterns on this execept for the decreasing calls on the 31st for the reason mentioned above. There is the slight cyclical behavior we saw earlier.

```python
hourlycount = data_311.groupby(['Hour']).agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                              'DOHMH':'sum','DOF':'sum' ,'TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
hourlycount.plot()
plt.title('Calls per hour')
plt.ylabel('Call Volume')
```

There is a huge spike in calls during midnight which makes it difficult to compare to the other data. Apparently all the housing dept calls all were made in midnight which probably means that they aren't tracking all their calls correctly. I am going to plot the calls from 1 am to the rest of the day to look at the trend without the outlier. This graph is skewed because it isn't true that all those calls happened on midnight. They just put midnight if they didn't record it. We can think of midnight is missing data. For a more accurate description, I am going to look at midnight separately and the rest of the day separately.

```python
midnight = hourlycount[:1]
del midnight['Total']
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
midnight.plot(kind = 'barh')
plt.title('Calls at Midnight')
plt.xlabel('Call Volume')
```

As you can see most of the calls at midnight are from the housing department, with negligable amounts from NYPD and DOHMH

```python
hr = hourlycount[1:]
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
hr.plot()
del hr['Total']
hr.plot()
plt.title('Calls per Hour')
plt.ylabel('Call Volume')
```

So I plotted the calls with and without the total to properly assess the calls. It seems like people are calling regarding DOT complaints right after their morning commute and right before their evening commute. This is probably because people are seeing complaints throughout the day and want to report it. For NYPD their calls seem to start at rush hour then stays stagnant throughout the day and increases at night. for the rest of the agencies they recieve their complaints around the 9-5 workday. This makes sense as that is around the time the agency employees are available. the housing department has 0 calls showing that they have a data storage error

```python
data_311['day_of_week'] = [date.dayofweek for date in data_311["Created Date"]]
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
data_311.pivot_table(values=['Total'],columns='day_of_week', aggfunc='count').plot(kind='bar')
plt.title('Calls per Day of the Week')
plt.ylabel('Call Volume')
```

This shows that half as many calls happen on weekends. Perhaps it is becuase offices are closed so it is more difficult to reach someone or because people aren't out as often on weekends so have less issues. However this clearly shows the cycle evident in the first line graph

```python
boroughcount = data_311.groupby('Borough').agg({ 'HPD':'sum', 'DOT':'sum','NYPD':'sum','DEP':'sum', 'DSNY':'sum', 'DOB':'sum','DPR':'sum',
                                              'DOHMH':'sum','DOF':'sum' ,'TLC':'sum','Total':'count'})
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
boroughcount.plot(kind = 'bar',)
plt.title('Calls per borough')
plt.ylabel('Call Volume')
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8, 6), dpi=80)
borough = boroughcount
del borough['Total']
borough.plot(kind = 'bar')
B =borough.T
B.plot(kind = 'bar', stacked = True)
plt.title('Call per borough')
plt.ylabel('Call Volume')
```

This graphs shows insight into NYC population. The fact that most TLC complaints happened in Manhattan shows that taxis are used most often in Manhattan. The order of the calls are for HPD show which boroughs' residents live in apartment buildings as opposed to private houses. The calls for DOT show how mouch each borough's residents drive instead of using public transportation. the fact that DOF is mostly unspecified and is the main contributer in that column shows that the DOF doesn't keep track of borough.
