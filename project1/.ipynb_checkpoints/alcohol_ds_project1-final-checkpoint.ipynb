#Merged dataset
Merging dataset1 & 2

```python
cancer_dataset_allcancer = cancer_dataset
sub_alcohol_data = alcohol_data

#Merged dataset
merged_all_cancer = pd.merge(cancer_dataset_allcancer, sub_alcohol_data, left_on=['State', 'Year'], right_on=['State', 'Year'], how='outer').dropna().sort(['Cancer Sites','State', 'Year'], ascending=[0,1, 1])
merged_all_cancer['cancer_percent'] = (merged_all_cancer['Count'] / merged_all_cancer['Population_over_14'])*100000000
merged_all_cancer['cancer_percent_original'] = (merged_all_cancer['Count'] / merged_all_cancer['Population_over_14'])
merged_all_cancer.head()
```


## Are you drinking this much?

### An extra beer a day could kill you? 


***Then we look at the top 2 states that had highest liver cancer rate and their drinking statistics (includes all types of alcohol).***   
* 5737461/2388982 = 2.4 gallons of ethanol per year per capita
* 2.4 gallons per year = 9085 grams per year
* 9085/8=1135 drinks per year, 8 grams of alcohol in a standard beer
* 1135/52=21 beers a week or ***3 beers*** in a given day.   

***We look at the bottom 2 states that had the lowest liver cancer rate and their drinking statistics.***    
* 2113016 / 1158477 = 1.8 gallons per year.  
* 1.8 gallons per year = 6813 grams per year.  
* 6813 / 8=851 drinks per year, 8 grams of alcohol in a standard beer.  
* 851 / 52=16 beers a week or ***2.2 beers*** in a given day.  

We see that the states that consistently have higher liver cancer rate also have higher total alcohol consumption, whether it's beer, wine, or spirits. The beer, wine, and spirits level vary within each state. But they all have one thing in common - higher total alcohol consumption, higher liver cancer rate.

Note: One unit is 10ml or 8g of pure alcohol. This equals one 25ml single measure of whisky (ABV 40%), or a third of a pint of beer (ABV 5-6%) or half a standard (175ml) glass of red wine (ABV 12%)


##**Conclusion**:  

It's not about whether you drink beer, spirits or wine, it's about the amount of ethanol that's present in the drink. 
If you are drinking around 2 beers a day or 16 a week. You may be safe.
If you are driking on average 3 beers a day or 19 a week, then you are probably hitting the liver cancer line.
